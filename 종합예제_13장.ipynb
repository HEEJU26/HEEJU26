```python
print("=== 결측치 개수 ===")
print(df.isnull().sum())
```

```python
total_count = len(df)

def missing_ratio(column):
    return (df[column].isnull().sum() / total_count) * 100
```

```python
print("=== 주요 컬럼 결측치 비율 ===")
print(f"면적 결측치 비율: {missing_ratio('면적'):.2f}%")
print(f"지역 결측치 비율: {missing_ratio('지역'):.2f}%")
print(f"개원일 결측치 비율: {missing_ratio('개원일'):.2f}%")
print(f"주요시설 결측치 비율: {missing_ratio('주요시설'):.2f}%")
```

```python
print(f"결측치 제거 전: {df.shape[0]}개 행")

# 필수 컬럼 결측치 제거
df_clean = df.dropna(subset=['지역', '면적'])

print(f"필수 정보 결측치 제거 후: {df_clean.shape[0]}개 행")
```

```python
df_clean = df_clean.drop_duplicates()
print(f"중복행 제거 후: {df_clean.shape[0]}개 행")
```

```python
print("=== 정제 후 결측치 개수 ===")
print(df_clean.isnull().sum())
```

```python
print(f"최종 데이터 크기: {df_clean.shape}")
```

```python
# 면적 통계
total_area = df_clean['면적'].sum()
mean_area = df_clean['면적'].mean()
median_area = df_clean['면적'].median()

print("=== 서울시 전체 공원 통계 분석 ===")
print(f"전체 공원 총 면적: {total_area:,.2f}㎡")
print(f"전체 공원 평균 면적: {mean_area:,.2f}㎡")
print(f"전체 공원 중앙값: {median_area:,.2f}㎡")

# 최소 면적 공원
min_park = df_clean[df_clean['면적'] == df_clean['면적'].min()]
print("\n최소 면적 공원:")
print(min_park[['공원명', '면적']])

# 최대 면적 공원
max_park = df_clean[df_clean['면적'] == df_clean['면적'].max()]
print("\n최대 면적 공원:")
print(max_park[['공원명', '면적']])
```

```python
top10 = df_clean.nlargest(10, '면적').reset_index(drop=True)

print("=== 면적 상위 10대 공원 ===")
for i in range(len(top10)):
    print(
        f"{i+1}위: {top10.loc[i, '공원명']} "
        f"({top10.loc[i, '지역']}) - {top10.loc[i, '면적']:,.2f}㎡"
    )
```

```python
large_parks = df_clean[df_clean['면적'] >= 500_000]
large_parks = large_parks.sort_values('면적', ascending=False).reset_index(drop=True)

print(f"면적 50만㎡ 이상 대형 공원 개수: {len(large_parks)}개\n")

print("=== 대형 공원 목록 ===")
for i in range(len(large_parks)):
    print(
        f"{i+1}. {large_parks.loc[i, '공원명']} "
        f"({large_parks.loc[i, '지역']}) - {large_parks.loc[i, '면적']:,.2f}㎡"
    )
```

```python
import plotly.express as px

# 1) 지역별 공원 개수 집계
region_count_df = df_clean.groupby('지역')['공원명'].count().reset_index()

# 2) 컬럼명 변경
region_count_df.columns = ['지역', '공원개수']

# 3) 상위 10개만 추출 (공원개수 기준 내림차순)
top10 = region_count_df.sort_values('공원개수', ascending=False).head(10)

# 4) 막대그래프 생성 (막대 위에 개수 표시)
fig = px.bar(
    top10,
    x='지역',
    y='공원개수',
    title='서울시 지역별 공원 개수 Top 10',
    text='공원개수',
    labels={'공원개수': '공원 개수(개)', '지역': '지역(자치구)'}
)

# 5) 텍스트를 막대 위에 표시
fig.update_traces(textposition='outside')

# 6) 그래프 출력
fig.show()
```
